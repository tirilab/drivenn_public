```python
def format_output(file_name):
    try:
        f_df = pd.read_csv(file_name)
    except:
        print(f'Oops, unable to find file {file_name}')
        return None
    
    roc_score, aupr_score, precision, recall, f_score, acc, mcc, duration = [], [], [], [], [], [], [], []

    for i in range(len(f_df)):
        curr_row = [float(val.strip().replace("[", "").replace("]", "").replace("'", "")) for val in f_df['scores'][i].split(",")[1::]]
        roc_score.append(curr_row[1])
        aupr_score.append(curr_row[2]) 
        precision.append(curr_row[3])
        recall.append(curr_row[4])
        f_score.append(curr_row[5])
        acc.append(curr_row[6])
        mcc.append(curr_row[7])
        
    formatted_df = pd.DataFrame({"se": [se2name[i] for i in f_df['ddi']], "se_id": list(f_df['ddi']),  "roc_score": roc_score, "aupr_score": aupr_score, "precision": precision, "recall": recall, "f_score": f_score, "acc": acc, "mcc": mcc})
    return formatted_df
```

```python
full_driven_df = pd.DataFrame()
for ep in e:
    prefix = '../data/cvd_output/cvd_output'
    if 'one_cvd' in ep:
        file_path = f'{prefix}_{ep}.csv'
    else:
        file_path = f'{prefix}_all_dps_pca_0.95_{ep}.csv'
        
    df = format_output(file_path)
    df['model'] = ['DrIVeNN'] * len(df)
    df['dataset'] = [ep] * len(df)
    full_driven_df = pd.concat([full_driven_df, df], axis = 0)
```

## Look at severity bins

```python
def f(score):
    sev_dict = {"low": (0,0.2), "minor": (0.2, 0.4), "moderate": (0.4, 0.6), "major": (0.6, 0.8), "critical": (0.8, 1)}
    
    if score < 0.2:
        return "low"
    elif score < 0.4:
        return "minor"
    elif score < 0.6:
        return "moderate"
    elif score < 0.8:
        return "major"
    else:
        return "critical"
```

```python
merged = se_df_cat[~se_df_cat['saedr_score'].isna()]
merged['severity_bin'] = [f(x) for x in merged.saedr_score]
```

```python
def g(score):
    if score > 0.95:
        return "> 0.95"
    elif score > 0.9:
        return "(0.9, 0.95)"
    elif score > 0.85:
        return "(0.85, 0.90)"
    elif score > 0.8:
        return "(0.8, 0.85)"
    else:
        return "< 0.8"
```

```python
merged['roc_bin'] = [g(x) for x in merged.roc_score]
merged['aupr_bin'] = [g(x) for x in merged.aupr_score]
```

```python
one_roc = merged[merged['dataset'] == 'one_cvd']\
.groupby(["roc_bin"])\
.agg({"se_name": list, "se_id": len,  "saedr_score": list})
```
